```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy import signal
import glob
```

```python
# load files into arrays

myFiles = glob.glob('*.txt')
S1Lb = np.loadtxt(myFiles[0])
S2LF = np.loadtxt(myFiles[1])
S2St = np.loadtxt(myFiles[2])
S1LF = np.loadtxt(myFiles[3])
S1LLF = np.loadtxt(myFiles[4])
S1RW = np.loadtxt(myFiles[5])
S2RF = np.loadtxt(myFiles[6])

myFiles
```

```python
# define domains
fs = 128
Ts = 1/fs
```

```python
# plot raw data

# subplots
fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(7, 1, figsize=(14, 10), sharex=True)
fig.suptitle('Velocity data')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1Lb
N0 = len(S1Lb)
t0 = np.linspace(0,N0*Ts,N0)
ax1.plot(t0, S1Lb)
ax1.set_title('S1-Lumbar')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1LF
N1 = len(S1LF)
t1 = np.linspace(0,N1*Ts,N1)
ax2.plot(t1, S1LF)
ax2.set_title('S1-LeftFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1LLF
N2 = len(S1LLF)
t2 = np.linspace(0,N2*Ts,N2)
ax3.plot(t2, S1LLF)
ax3.set_title('S1-LowerLeftLeg')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1RW
N3 = len(S1RW)
t3 = np.linspace(0,N3*Ts,N3)
ax4.plot(t3, S1RW)
ax4.set_title('S1-RightWrist')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2LF
N4 = len(S2LF)
t4 = np.linspace(0,N4*Ts,N4)
ax5.plot(t4, S2LF)
ax5.set_title('S2-LeftFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2St
N5 = len(S2St)
t5 = np.linspace(0,N5*Ts,N5)
ax6.plot(t5, S2St)
ax6.set_title('S2-Sternum')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2RF
N6 = len(S2RF)
t6 = np.linspace(0,N6*Ts,N6)
ax7.plot(t6, S2RF)
ax7.set_title('S2-RightFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

plt.tight_layout()
plt.show()
```

```python
# fft of raw data signal 1 (S1Lb)

X1 = scipy.fft.fft(S1Lb[:N0])
X_mag = np.absolute(X1)
f = np.linspace(0, fs, N0)
plt.figure(figsize=(14, 5))
plt.title('FFT of S1Lb')
plt.plot(f[:int(N0/2)], 20*np.log10(X_mag[:int(N0/2)]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('magnitude (dB)')
```

```python
# use LPF with cutoff of 0.0005 Hz to detect slow turns only

h = signal.firwin(2001,cutoff=.0005,width=11,window='hamming',pass_zero='lowpass',fs=fs)
y0 = np.convolve(S1Lb,h,'same')

# plot fft of filtered signal 1 (S1Lb)
Y0 = scipy.fft.fft(y0[:N0])
Y_mag = np.absolute(Y0)
f = np.linspace(0, fs, N0)
plt.figure(figsize=(14, 5))
plt.title('FFT of filtered S1Lb')
plt.plot(f[:int(N0/2)], 20*np.log10(Y_mag[:int(N0/2)]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('magnitude (dB)')
```

```python
# plot filtered data

# subplots
fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(7, 1, figsize=(14, 10), sharex=True)
fig.suptitle('filtered data')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1Lb
y0 = np.convolve(S1Lb,h,'same')
ax1.plot(t0, y0)
ax1.set_title('S1-Lumbar')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1LF
y1 = np.convolve(S1LF,h,'same')
ax2.plot(t1, y1)
ax2.set_title('S1-LeftFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1LLF
y2 = np.convolve(S1LLF,h,'same')
ax3.plot(t2, y2)
ax3.set_title('S1-LowerLeftLeg')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1RW
y3 = np.convolve(S1RW,h,'same')
ax4.plot(t3, y3)
ax4.set_title('S1-RightWrist')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2LF
y4 = np.convolve(S2LF,h,'same')
ax5.plot(t4, y4)
ax5.set_title('S2-LeftFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2St
y5 = np.convolve(S2St,h,'same')
ax6.plot(t5, y5)
ax6.set_title('S2-Sternum')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2RF
y6 = np.convolve(S2RF,h,'same')
ax7.plot(t6, y6)
ax7.set_title('S2-RightFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

plt.tight_layout()
plt.show()
```

```python
# turn detection for S1Lb
# take the first backwards difference of y0 (S1Lb)

dy0 = [0.0]*len(y0)
trig_index = []

for x in range(N0):
    if x == 0:
        dy0[x] = y0[x]
    else:
        dy0[x] = y0[x] - y0[x-1]

# low pass filter the difference
dy0 = np.convolve(dy0,h,'same')

# pick out the values above threshold (0.0015 rad/s^2)
for x in range(N0):
    if np.abs(dy0[x]) > 0.0015:
        trig_index.append(x)
```

```python
# get range of threshold instead of individual points

min = []
max = []
min.append(trig_index[0])

for k in range(len(trig_index)):
    if k > 1 and k < len(trig_index)-1:
        if trig_index[k] != trig_index[k+1]-1:
            max.append(trig_index[k])
        if trig_index[k] != trig_index[k-1]+1:
            min.append(trig_index[k])
max.append(trig_index[len(trig_index)-1])
        
# plot values for validation
fig= plt.figure(figsize=(14,5))
plt.title('Backwards difference of turn velocity for S1Lb')
plt.xlabel('time (s)')
plt.ylabel('radians/sec^2')
plt.plot(t0,dy0)

for m in min:
    plt.axvline(x=m*Ts, color='r')
for n in max:
    plt.axvline(x=n*Ts, color='g')
    
plt.show()
```

```python
# find the average turn duration of S1 using Lumbar sensor data
duration0 = []
for p in range(len(min)):
    duration0.append(max[p]-min[p])
    
avg_dur0 = np.mean(duration0)
avg_dur0*Ts
```
