```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy
import glob
from scipy import signal
```

```python
# load files into arrays

# define domains
fs = 128
Ts = 1/fs

myFiles = glob.glob('*.txt')
S1Lb = np.loadtxt(myFiles[0])
S2LF = np.loadtxt(myFiles[1])
S2St = np.loadtxt(myFiles[2])
S1LF = np.loadtxt(myFiles[3])
S1LLF = np.loadtxt(myFiles[4])
S1RW = np.loadtxt(myFiles[5])
S2RF = np.loadtxt(myFiles[6])

N0 = len(S1Lb)
t0 = np.linspace(0,N0*Ts,N0)
dict0 = {'name': myFiles[0], 'values': S1Lb, 'length': N0, 'time': t0}

N1 = len(S1LF)
t1 = np.linspace(0,N1*Ts,N1)
dict1 = {'name': myFiles[1], 'values': S1LF, 'length': N1, 'time': t1}

N2 = len(S1LLF)
t2 = np.linspace(0,N2*Ts,N2)
dict2 = {'name': myFiles[4], 'values': S1LLF, 'length': N2, 'time': t2}

N3 = len(S1RW)
t3 = np.linspace(0,N3*Ts,N3)
dict3 = {'name': myFiles[5], 'values': S1RW, 'length': N3, 'time': t3}

N4 = len(S2LF)
t4 = np.linspace(0,N4*Ts,N4)
dict4 = {'name': myFiles[1], 'values': S2LF, 'length': N4, 'time': t4}

N5 = len(S2St)
t5 = np.linspace(0,N5*Ts,N5)
dict5 = {'name': myFiles[2], 'values': S2St, 'length': N5, 'time': t5}

N6 = len(S2RF)
t6 = np.linspace(0,N6*Ts,N6)
dict6 = {'name': myFiles[6], 'values': S2RF, 'length': N6, 'time': t6}

signals = [dict0,dict1,dict2,dict3,dict4,dict5,dict6]

myFiles
```

```python
# plot raw data

# subplots
fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7) = plt.subplots(7, 1, figsize=(14, 10), sharex=True)
fig.suptitle('Velocity data')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1Lb
N0 = len(S1Lb)
t0 = np.linspace(0,N0*Ts,N0)
ax1.plot(t0, S1Lb)
ax1.set_title('S1-Lumbar')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1LF
N3 = len(S1LF)
t3 = np.linspace(0,N3*Ts,N3)
ax2.plot(t3, S1LF)
ax2.set_title('S1-LeftFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1LLF
N4 = len(S1LLF)
t4 = np.linspace(0,N4*Ts,N4)
ax3.plot(t4, S1LLF)
ax3.set_title('S1-LowerLeftLeg')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S1RW
N5 = len(S1RW)
t5 = np.linspace(0,N5*Ts,N5)
ax4.plot(t5, S1RW)
ax4.set_title('S1-RightWrist')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2LF
N1 = len(S2LF)
t1 = np.linspace(0,N1*Ts,N1)
ax5.plot(t1, S2LF)
ax5.set_title('S2-LeftFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2St
N2 = len(S2St)
t2 = np.linspace(0,N2*Ts,N2)
ax6.plot(t2, S2St)
ax6.set_title('S2-Sternum')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

# plot S2RF
N6 = len(S2RF)
t6 = np.linspace(0,N6*Ts,N6)
ax7.plot(t6, S2RF)
ax7.set_title('S2-RightFoot')
plt.xlabel('time (s)')
plt.ylabel('rotation (rad/s)')

plt.tight_layout()
plt.show()
```

```python
X1 = scipy.fft.fft(S1Lb[:N0])
X_mag = np.absolute(X1)        # spectral magnitude
f = np.linspace(0, fs, N0)     # frequency variable
plt.figure(figsize=(14, 5))
plt.title('FFT of S1Lb')
plt.plot(f[:int(N0/2)], 20*np.log10(X_mag[:int(N0/2)])) # magnitude spectrum
plt.xlabel('Frequency (Hz)')
plt.ylabel('magnitude (dB)')
#plt.vlines(2.5,-25,60,colors='r',label='tap freq = 2.5Hz')
#plt.legend()
```

```python
h = signal.firwin(101,cutoff=5,width=11,window='hamming',pass_zero='lowpass',fs=fs)
y1 = np.convolve(S1Lb,h,'same')

Y1 = scipy.fft.fft(y1[:N0])
Y_mag = np.absolute(Y1)        # spectral magnitude
f = np.linspace(0, fs, N0)     # frequency variable
plt.figure(figsize=(14, 5))
plt.title('FFT of filtered S1Lb')
plt.plot(f[:int(N0/2)], 20*np.log10(Y_mag[:int(N0/2)])) # magnitude spectrum
plt.xlabel('Frequency (Hz)')
plt.ylabel('magnitude (dB)')
```
