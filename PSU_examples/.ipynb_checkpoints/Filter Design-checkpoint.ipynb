# Filter Design

## Overview
* Filters are a class of LTI systems
* This section focuses on __frequency-selective filters__
     * Designed to pass some frequencies
     * Stop other frequencies
     * Please keep in mind this isn't always the goal
     * Many applications wish to emphasize some time-domain feature 
         * Smoothing
         * Trend
         * Noise reduction
         * Channel inversion (equalization)
         * Estimation
         * Feature detection
     * Design approach in these cases may be very different
         
* Focus is on causal filters
     * Non-causal filters can often be designed and implemented from causal designs
* Goal: determine parameter values of transfer function or difference equation that meets specifications

### Often used for DT Processing of CT Systems

Advantages
* Cost
* Complexity
* Repeatability
* Flexibility
    * Linear phase 
    * Sometimes non-causal
    * FIR
    * Pure delay
    * Nonlinear
* But downside is additional delay 
    * Reconstruction of theoretical D->A is noncausal

### Filter Types
* IIR
    * Rational function of $z$
    * Based on mapping continuous-time designs
* FIR
    * Polynomial in $z$
    * Designs based on windowing and iterative algorithms
    
### Design Stages
1. Specification of properties
    * Depends on application
    * Often job of engineer must be from user requirements (often vague) to specifications
2. Approximation of specifications with DT design
3. Realization of the system
    * Details depend on implementation
    * Quantization of coefficients may matter---particularly for poles near unit circle
    * See Chapter 6

## Transforms Review

### Discrete-Time Fourier Transform
\begin{align*}
    X(e^{j\omega}) &= \sum_{n=-\infty}^{\infty} x[n] e^{-j\omega n} \\
    x[n]           &= \frac{1}{2\pi} \int_{-\pi}^{\pi} X(e^{j\omega}) e^{j\Omega n} d\omega
\end{align*}

### Continuous-Time Fourier Transform
\begin{align*}
    X(j\Omega) &= \int_{-\infty}^{\infty} x(t) e^{-j\Omega t} dt \\
    x(t)       &= \frac{1}{2\pi }\int_{-\infty}^{\infty} X(j\Omega) e^{j\Omega t} d\Omega
\end{align*}

### Laplace Transform
\begin{align*}
    X(s) &= \int_{0^{-}}^{\infty} x(t) e^{-s t} dt \\
    x(t) &= \frac{1}{2\pi j}\int_{\sigma-j\infty}^{\sigma+j\infty} X(x) e^{s t} ds
\end{align*}
such that $\sigma + j\omega$ is in the region of convergence (ROC)

## Filter Specifications (7.1)

* Will primarily focus on lowpass filters
    * Can modulate lowpass response to other frequencies to create other filter types (see Section~7.4)
* Draw diagram of typical tolerance limits
    * Maximum passpand gain: $1+\delta_{p1}$
    * Minimum passpand gain: $1-\delta_{p2}$
    * Maximum stopband gain: $\delta_s$
    * Passband edge frequency: $\omega_p$
    * Stopband edge frequency: $\omega_s$
    * Transition band may be unconstrained or constrained by maximum passband gain
* Phase response
    * IIR: no requirements
    * FIR: Generalized linear phase
* Other requirements
    * Stable
    * Causal
* Gains often expressed in units of dB: $20 \log_{10} \delta$

### CT and DT Relationship
* Suppose DT system used for CT problem
* The DT system can't impact frequencies above half the sample rate (sampling theorem)

\begin{equation*}
H_{e}(j\Omega) = 
\begin{cases}
    H(e^{j\Omega T}) & |\Omega|<\pi/T \\
    0 & |\Omega|\geq\pi/T \\
\end{cases}
\end{equation*}



## IIR Filter Design from CT IIR Filters

* IIR filter design is well understood and advanced
* Many IIR designs have closed-form design formulas
* Cannot directly apply same methods in discrete time
    * No simple closed-form design formulas
    * Unlike CT, frequency response of DT filters is periodic

### CT Filters and CT Processing
* DT filters can be used to process CT signals (with sampling and bandlimited interpolation)
* We are using CT filters in a completely different way
* We are using them to create DT filters
* Three steps
    * Convert DT specifications to CT 
    * Create CT filter
    * Transform CT filter to DT
* Essential properties of CT filter must survive conversion to DT
* Implies imaginary axis of $s$-plane must map to unit circle of $z$-plane
* Stable CT filter must lead to a stable DT filter
    * Left half of $s$-plane must map to interior of unit circle of the $z$-plane
    * Poles in left half $s$-plane must land inside the unit circle of the $z$-plane


###  Filter Design by Impulse Invariance (7.2.1)

Possible filter objectives
* Same impulse response (sampled)
* Maintain characteristics such as overshoot, energy compaction, time-domain ripple, etc.
* Same frequency response over a band of frequencies


#### Impulse Invariance Design

\begin{equation*}
    h[n] = T_d h_c(n T_d)
\end{equation*}
where $T_d$ represents the sampling interval and $h_c(t)$ is the continuous-time impulse response

* $T_d$ is totally unimportant
* It is used twice in the process
    * Convert DT specifications to CT
    * Transform CT design to DT
* The effects of $T_d$ in these two steps cancels out
* If processing a continuous-time signal with a DT system, $T_d$, is totally independent and separate from the sampling interval

I assume you learned about sampling in a previous class. You should have learned from that class that if a signal is sampled (such as the impulse response above), the resulting relationship of the transforms is as follows
\begin{equation*}
    H(e^{j\omega}) = \sum_{k=-\infty}^{\infty} H_c\left(j \frac{\omega}{T_d} + j \frac{2\pi}{T_d} k\right)
\end{equation*}

__Diagram__
* Draw diagram of CT and DT frequency responses
* Unity gain at DC
* Show effects of aliasing

If CT filter is bandlimited such that
\begin{equation*}
    H_c(j\Omega) = 0 \text{ for } |\Omega| \geq \pi/T_d
\end{equation*}
Then
\begin{equation*}
    H(e^{j\omega}) = H_c\left(j \frac{\omega}{T_d}\right) \text{ for } |\omega|\leq \pi
\end{equation*}

* Practical CT filters are not bandlimited
* Some aliasing will occur
* However if gain is small at high frequencies, aliasing effect may be tolerable for the application
* Good idea to use CT filters with no stopband ripple, when possible

__Transforming Specifications__

\begin{equation*}
    \Omega = \omega/T_d
\end{equation*}

* Again $T_d$ doesn't much matter
* Cannot be used to control aliasing
* If sampling rate is increased, continuous-time cutoff frequency will increase just as much

__Obtaining $H(z)$__

Once we have our CT filter $H_c(s)$, we obtain $H(z)$

Suppose poles are distinct and we've performed partial fraction expansion such that
\begin{equation*}
    H_c(s) = \sum_{k=1}^{N} \frac{A_k}{s-s_k}
\end{equation*}

CT Impulse response is then
\begin{equation*}
    h_c(t) = \sum_{k=1}^{N} A_k e^{s_k t} u(t)
\end{equation*}
and DT impulse response is
\begin{align*}
    h[n] &= T_d \sum_{k=1}^{N} A_k e^{s_k n T_d} u[n]\\
         &= T_d \sum_{k=1}^{N} A_k (e^{s_k T_d})^n u[n]
\end{align*}
so the system function is then
\begin{equation*}
    H(z) = \sum_{k=1}^{N} \frac{T_d A_k}{1-e^{s_k T_d} z^{-1}}
\end{equation*}

* Pole at $s_k$ transformed to pole at $z=e^{s_k T_d}$
* Coefficients of partial fraction expansion same to within a scale factor of $T_d$
*  If CT pole at $s_k$ is in the left half plane, then the DT pole will be in the unit circle
*  DT filter is stable of CT filter is stable
*  Not a simple mapping from $s$-plane to $z$-plane, however
*  Zeros are mapped differently than poles

__Key Points__
* Relationship of CT and DT frequencies is linear
    * Except for effect of aliasing
* Only appropriate for bandlimited filters
* Does not work (directly) for highpass or bandstop filters
    * High frequency passband would be aliased in discrete time

### Bilinear Transformation (7.2.2)
* Unlike impulse invariance, maps the entire $s=j\Omega$ axis to one revolution of unit circle
* Necessarily nonlinear 
    * Maps infinite span to finite span
* Restricts applicability to situations where nonlinear mapping is acceptable
    * For example, filters with bands of nearly constant gain
    
\begin{align*}
    s &= \frac{2}{T_d} \left(\frac{1-z^{-1}}{1+z^{-1}}\right) \\
    H(z) &= H_c\left(\frac{2}{T_d} \left(\frac{1-z^{-1}}{1+z^{-1}}\right)\right)
\end{align*}

* As with impulse invariance, $T_d$ does not matter
* If we start with discrete time specifications, the effect of $T_d$ in determining the continuous-time specifications will cancel the effect of mapping the filter back to discrete-time
* There is also a relationship between DT difference equations and CT differential equations,  but it is not important
     * See text for details

\begin{align*}
    z &= \frac{1+(T_d/2)s}{1-(T_d/2)s}
\end{align*}
and when $s=\sigma + j\Omega$
\begin{align*}
    z &= \frac{1+\sigma T_d/2 + j\Omega T_d/2}{1-\sigma T_d/2 + j\Omega T_d/2}
\end{align*}
* If $\sigma<0$ then $|z|<1$
* if $\sigma>0$ then $|z|>1$

On the unit circle $z=e^{j\omega}$ we can show
\begin{align*}
    \Omega &= \frac{2}{T_d} \tan(\omega/2) \\
    \omega &= 2 \arctan(\Omega T_d/2)
\end{align*}

* Sketch (see Figure 7.7)
* Horizontal axis $\Omega$
* Vertical axis $\omega$
* Saturates at $\pi$ and $-\pi$

Key Points
* No aliasing (unlike impulse invariance)
* But nonlinear compression of CT frequency axis
    * Also warps the phase response
    * If the CT filter has linear phase, the DT filter will not
* Piecewise constant magnitude filters okay

## 7.3 Common DT IIR Filters

Four common types of IIR Filters
* __Butterworth__
    * Monotonic response (no ripple)
    * Generally the highest order
    * Maximally flat (derivatives are 0 at $\omega=0$)
* __Chebyshev Type I__
    * Equiripple response in passband
    * Monotonic stopband
* __Chebyshev Type II__
    * Monotonic passband
    * Equiripple stopband
* __Elliptic__
    * Equiripple passband
    * Equiripple stopband
    * Lowest order

* Filter properties preserved by bilinear transformation
* The mapping of specifications to CT, CT design, and transformation of CT filter to DT generally handled by single scripts in Python and Matlab

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
def PlotSystem(sos,Rs=None,Rp=None,wp=None,ws=None):
    w, h = signal.sosfreqz(sos,2000)

    (zeros,poles,k) = signal.sos2zpk(sos)

    #fig = plt.figure()
    figure = plt.figure(num=1,figsize=(10,6),dpi=100)
    figure.clf()
    axes = figure.add_subplot(3,2,1)
    for zeroAngle in np.angle(zeros):
        axes.axvline(zeroAngle,color='b',linestyle=':',linewidth=0.5)
    axes.plot(w, 20 * np.log10(np.maximum(abs(h), 1e-5)),color='r')
    if Rs:
        yMin = -Rs-20
    else: 
        yMin = -60-20
    yMax = 5.0
    if Rs is not None and ws is not None:
        axes.fill_between([0,ws[0]*np.pi],[-Rs,-Rs],[yMax,yMax],linewidth=0.0,facecolor='k',alpha=0.2)        
        axes.fill_between([ws[1]*np.pi,np.pi],[-Rs,-Rs],[yMax,yMax],linewidth=0.0,facecolor='k',alpha=0.2)
    if Rp is not None and wp is not None:
        ap = 10.0**(Rp/(-20.0))
        axes.fill_between(wp*np.pi,[yMin,yMin],[ap,ap],linewidth=0.0,facecolor='k',alpha=0.2)
        axes.fill_between(wp*np.pi,[1.0,1.0],[yMax,yMax],linewidth=0.0,facecolor='k',alpha=0.2)
    axes.set_ylabel('Amplitude (dB)')
    axes.set_ylim([yMin,yMax])
    axes.set_xlim([0.0,np.pi])

    axes = figure.add_subplot(3,2,2)
    axes.plot(w,abs(h),color='r')
    yMin = 0.0
    yMax = 1.05
    for zeroAngle in np.angle(zeros):
        axes.axvline(zeroAngle,color='b',linestyle=':',linewidth=0.5)
    if Rs is not None and ws is not None:
        axes.fill_between([0,ws[0]*np.pi],[-Rs,-Rs],[yMax,yMax],linewidth=0.0,facecolor='k',alpha=0.2)
        axes.fill_between([ws[1]*np.pi,np.pi],[-Rs,-Rs],[yMax,yMax],linewidth=0.0,facecolor='k',alpha=0.2)
    if Rp is not None and wp is not None:
        ap = 10.0**(Rp/(-20.0))
        axes.fill_between(wp*np.pi,[yMin,yMin],[ap,ap],linewidth=0.0,facecolor='k',alpha=0.2)
        axes.fill_between(wp*np.pi,[1.0,1.0],[yMax,yMax],linewidth=0.0,facecolor='k',alpha=0.2)
    axes.set_ylabel('Amplitude')
    axes.set_ylim([0.0,1.05])
    axes.set_xlim([0.0,np.pi])

    axes = figure.add_subplot(3,2,3)
    for angle in np.arange(-360.0,360.0+1,180.0):
        axes.axhline(angle,color='k',linestyle=':',linewidth=0.5)
    for zeroAngle in np.angle(zeros):
        axes.axvline(zeroAngle,color='b',linestyle=':',linewidth=0.5)    
    axes.plot(w, np.angle(h)*(180.0/np.pi),color='g')
    axes.set_ylabel('Phase ($^o$)')
    axes.set_ylim([-190.0,190.0])
    axes.set_xlim([0.0,np.pi])

    axes = figure.add_subplot(3,2,4)
    phaseUnwrapped = np.unwrap(np.angle(h))*(180.0/np.pi)
    for angle in np.arange(-360.0*10,360.0*10+1,180.0):
        axes.axhline(angle,color='k',linestyle=':',linewidth=0.5)
    for zeroAngle in np.angle(zeros):
        axes.axvline(zeroAngle,color='b',linestyle=':',linewidth=0.5)
    axes.plot(w,phaseUnwrapped,color='g')
    axes.set_ylabel('Phase ($^o$)')
    axes.set_xlim([0.0,np.pi])
    axes.set_ylim(phaseUnwrapped.min(),phaseUnwrapped.max())

    (num,den) = signal.sos2tf(sos)
    wgd,groupDelay = signal.group_delay((num,den), w=2**13)
    groupDelay[0] = 0.0
    axes = figure.add_subplot(3,2,5)
    for zeroAngle in np.angle(zeros):
        axes.axvline(zeroAngle,color='b',linestyle=':',linewidth=0.5)
    axes.axhline(0.0,color='k',linestyle=':',linewidth=0.5)
    axes.plot(wgd, groupDelay)
    axes.set_ylabel('GD ($n$)')
    yMin = np.percentile(groupDelay,0.05)-1.0
    yMax = np.percentile(groupDelay,99.5)+1.0
    axes.set_ylim([yMin,yMax])
    axes.set_xlim([0.0,np.pi])

    axes = figure.add_subplot(3,2,6)
    (num,den) = signal.sos2tf(sos)
    ni = np.arange(150)
    xi = signal.unit_impulse(ni.shape)
    hi = signal.lfilter(num,den,xi)
    axes.axhline(0.0,color='k',linestyle=':',linewidth=0.5)
    axes.vlines(ni,0,hi,color='b')
    axes.set_ylabel('h[n]')
    axes.set_xlim([0.0,ni[-1]+0.5])

    figure = plt.figure(num=2,figsize=(4,4),dpi=100)
    figure.clf()

    axes = figure.add_subplot(1,1,1)
    t = np.linspace(0,np.pi*2,100)
    axes.plot(np.cos(t), np.sin(t),color='k',linewidth=0.8,linestyle=':')
    axes.axhline(0.0,color='k',linewidth=0.8,linestyle=':')
    axes.axvline(0.0,color='k',linewidth=0.8,linestyle=':')
    axes.plot(np.real(zeros),np.imag(zeros),marker='o',linestyle='none',markerfacecolor='None',color='b',linewidth=0.1)
    axes.plot(np.real(poles),np.imag(poles),marker='x',linestyle='none',color='r',linewidth=0.1)
    axes.set_xlim([-1.2,1.2])
    axes.set_ylim([-1.2,1.2])
    axes.set_aspect('equal')
    axes.set_xlabel('Real')
    axes.set_ylabel('Imaginary')
    axes.axis('off')
```

```python
wp = np.array([0.30,0.50]) # Where 1.0 = pi radians/sample
ws = np.array([0.25,0.55]) # Where 1.0 = pi radians/sample
ap = 0.95
Rs = 60 # Stopband attenuation maximum
Rp = -20.0*np.log10(ap)  # Maximum ripple in the passband. (dB)

(ord,wn) = signal.ellipord(wp, ws, Rp, Rs)
sos = signal.iirfilter(ord,wn,rp=Rp,rs=Rs,btype='band',ftype='ellip',output='sos')

PlotSystem(sos,Rp=Rp,Rs=Rs,wp=wp,ws=ws)
```

## Frequency Transformations (7.4)

* We can transform lowpass filters into other types 
* Works in either DT or CT
* With bilinear transform, could do this in DT
* Won't work with impulse invariance because of aliasing
* With either type, can transform in DT

Similar to the bilinear transform, the DT-to-DT transforms can be expressed as a substitution for $z^{-1}$ with
\begin{equation*}
    Z^{-1} = G(z^{-1})
\end{equation*}
such that
\begin{equation*}
    H(z) = \left. H_{\mathrm{lp}}(Z) \right|_{Z^{-1}=G(z^{-1})}
\end{equation*}

* $G(z^{-1})$ must be a rational function of $z^{-1}$
* Inside of unit circle of $Z$ plane must map to inside of unit circle of $z$ plane
* Unit circle of $Z$-plane must map to unit circle of $z$-plane

* There are algebraic expressions for transformations for the four common filter types (Table 7.1)
* Frequency modulation is another possible approach

## FIR Filter Design by Windowing (7.5)

* FIR filters are designed in discrete time
* Based on either 
    * Desired impulse response
    * Desired frequency response
* Simplest method called the __window method__


Suppose you have an ideal filter response
\begin{equation*}
    H_d(e^{j\omega}) = \sum_{n=-\infty}^{\infty} h_d[n] e^{-j \omega n}
\end{equation*}
where $h_d[n]$ is the corresponding impulse response
\begin{equation*}
    h_d[n] = \frac{1}{2\pi} \int_{-\pi}^{\pi} H_d(e^{j\omega}) e^{j\omega n} d\omega 
\end{equation*}

* Most idealized systems are piecewise constant or piecewise smooth
* Discontinuities at boundaries
* Impulse responses are therefore often
    * Infinitely long
    * Non-causal

Recall that we can create a truncated impulse response by multiplication of a finite-duration window $w[n]$
\begin{equation*}
    h[n] = h_d[n] w[n]
\end{equation*}
where $w[n]=0$ for $n<M_1$ and $n>M_2$, say where $M = M_2 - M_1 + 1$.

In the frequency domain this means
\begin{equation*}
    H(e^{j\omega}) = \frac{1}{2\pi} \int_{-\pi}^{\pi} H_d(e^{j\theta}) W(e^{j(\omega-\theta)}) d\theta
\end{equation*}

* Periodic convolution in frequency domain
* Smeared or smoothed version of the desired response
* Ideally we would like $W(e^{j\omega})$ to be an impulse train


### Window Design Goals
* Minimize window duration ($M+1$)
* $W(e^{j\omega})$ as concentrated as possible at $\omega=\ell 2\pi$ where $\ell$ is an integer
    * Narrow main lobe
    * Short side lobes
    

__Primary Tradeoff__
Main lobe width vs. side lobe height
* Creates oscillations near discontinuities
* Related to Gibbs phenomenon (truncated Fourier series)


### Common Windows (7.5.1)
* Rectangular
    * $+$ Smallest error from ideal
    * $+$ Narrowest main lobe
    * $-$ Highest side lobe (-13~dB)
* Bartlett (Triangular)
    * Essentially never used for filter design
* Hann (Hanning)
* Hamming
    * Side lobes are nearly uniform in height
    * Similar to a particular type of Kaiser window
* Blackman
    * $-$ Widest main lobe
    * $+$ Smallest side lobes
    
### Window Approach to FIR Design
* $+$ Fast (can design on the fly)
* $+$ Deterministic
* $-$ No control over how it meets or doesn't meet specifications
* $-$ No control over ripple in passband vs stopband (no weighting)

```python
import scipy

M = 20 # Filter order

a = np.array([1.0])
b = scipy.signal.get_window('boxcar',M)
b = b/sum(b)
sos = scipy.signal.tf2sos(b,a)
PlotSystem(sos)
```

```python
b = scipy.signal.get_window('hann',M)
b = b/sum(b)
sos = scipy.signal.tf2sos(b,a)
PlotSystem(sos)
```

```python
b = scipy.signal.get_window('hamming',M)
b = b/sum(b)
sos = scipy.signal.tf2sos(b,a)
PlotSystem(sos)
```

```python
b = scipy.signal.get_window('blackman',M)
b = b/sum(b)
sos = scipy.signal.tf2sos(b,a)
PlotSystem(sos)
```

## Optimal FIR Design

* $+$ FIR filters are great because can easily obtain linear phase
* $-$ Not so great only because may take a lot of taps (clock cycles)
* $+$ Can quickly design on the fly with windowing
* If they have even symmetry, we have another opportunity
