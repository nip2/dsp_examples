#  Linear Constant-Coefficient Difference Equations (2.5)

## Definition
Systems for which the output $y[n]$ satisfies the $N$th order linear constant-coefficient difference equation of the form

$\sum_{k=0}^{N} a_k y[n-k] = \sum_{m=0}^{M} b_m x[n-m]$

Example:

$\begin{align*}
    y[n] &= \sum_{k=-\infty}^{n} x[k] \\
         &= x[n] + y[n-1]
\end{align*}$

* As shown by example, representation of LTI systems is not unqiue
* Practical considerations may determine the best representation


## Uniqueness
Linear constant-coefficient difference equation does not uniquely determine the output

$y[n] = y_p[n] + y_h[n]$

where $y_h[n]$ is any solution such that for $x[n]=0$

$\sum_{k=0}^{N} a_k y_h[n-k] = 0$

Called the __homogenous difference equation__. 

The solution is of the form

$y_h[n] = \sum_{m=1}^{N} A_m z_m^n$

where the coefficients $A_m$ are chosen to satisfy a set of auxiliary conditions. The complex numbers $z_m$ are the roots of the polynomial

$A(z) = \sum_{k=0}^{N} a_k z^{-k}$

* A set of $N$ auxiliary conditions are required to uniquely specify $y[n]$
* Often specified in the form of $y[-1], y[-2], \dots, y[-N]$

Then

$y[n] = -\sum_{k=1}^{N} \frac{a_k}{a_0} y[n-k] + \sum_{k=0}^{M} \frac{b_k}{a_0} x[n-k]$

Similarly

$y[n-N] = -\sum_{k=0}^{N-1} \frac{a_k}{a_N} y[n-k] + \sum_{k=0}^{M} \frac{b_k}{a_0} x[n-k]$

If a system is 

* Characterized by a linear constant-coefficient difference equation
* Time invariant
* Causal

then the solution $y[n]$ is unique

Auxiliary conditions are often stated as __initial-rest__. This means that if $x[n]=0$ for all $n<n_0$ then $y[n]=0$ for $n<n_0$ for some $n_0$. Sufficient to determine $y[n]$ for $n\geq n_0$

### Relationship to LTI Systems

* All LCCD equations at initial rest are LTI systems
* Not all LTI systems can be represented as LCCD
* However, most of the systems that we create (design), can be represented as LCCD systems


### FIR Systems
If $N=0$, then $y[n]$ is not a recursive equation and no auxiliary conditions are needed

$y[n] =  \sum_{k=0}^{M} \frac{b_k}{a_0} x[n-k]$

which means

$h[n] = \sum_{k=0}^{M} \frac{b_k}{a_0} \delta[n-k]$

Impulse response is clearly of finite duration
