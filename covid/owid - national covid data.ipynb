```python
# setup dataframe with world covid data (retreived 12-14-2020)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
#%matplotlib nbagg
```

```python
df = pd.read_csv('owid-covid-data.csv', parse_dates=True)
index = range(1,len(df))
us = df[df['location'].str.match('United States')]
```

```python
us
```

```python
#set date as index
us.set_index('date',inplace=True)

#plot data
fig, ax = plt.subplots(figsize=(15,7))
us['total_cases'].plot(ax=ax)

#set ticks every week
ax.xaxis.set_major_locator(mdates.MonthLocator())
#set major ticks format
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

ax.set_title('Total cases in US')
ax.set_ylabel('number of cases (10\'s of millions)')
```
