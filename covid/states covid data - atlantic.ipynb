```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
#%matplotlib nbagg
```

```python
# setup dataframe with US covid data (retreived 12-14-2020)

df = pd.read_csv('all-states-history.csv', parse_dates=True)
index = range(1,len(df))

df = df.sort_values(by='date') 
```

```python
df
```

```python
# select state and plot
state = 'FL'
column = 'hospitalized'

sdata = df[df['state'].str.match(state)]

#set date as index
sdata.set_index('date',inplace=True)

#plot data
fig, ax = plt.subplots(figsize=(15,7))
sdata[column].plot(ax=ax)

#set ticks every week
ax.xaxis.set_major_locator(mdates.MonthLocator())
#set major ticks format
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

ax.set_title('%s in %s' %(column, state))
ax.set_ylabel('number of people')
```
